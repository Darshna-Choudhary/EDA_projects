# Q2. Top 10 most polluted stations

```python
per_station_pollution = df.groupby("station")["pollutant_avg"].mean().reset_index().sort_values("pollutant_avg", ascending = False).head(10)
per_station_pollution
```

```python
plt.figure(figsize=(6, 8))
plt.plot(per_station_pollution["station"], per_station_pollution["pollutant_avg"])
plt.xlabel("Stations")
plt.ylabel("Pollution score")
plt.title("Top 10 polluted stations")
plt.xticks(rotation = 80)
plt.tight_layout()
```

Since each station measures multiple pollutants, a single pollution value does not exist by default.
To fairly compare stations, I first computed the average value of each pollutant per station.
Then, I calculated the mean of these pollutant averages to obtain a composite pollution score per station.
Stations were ranked based on this score to identify the top 10 most polluted stations.
- This approach assumes equal weight for all pollutants and does not account for health-based pollutant thresholds.

# Q3. Compare pollution patterns across pollutants

```python
state_pollutant = df.groupby(["state", "pollutant_id"])["pollutant_avg"].mean().reset_index()
state_pollutant["normalised"] = (state_pollutant["pollutant_avg"] / state_pollutant.groupby("state")["pollutant_avg"].transform(sum))
state_pollutant
```

```python
pivot_df = state_pollutant.pivot(index = "state", columns = "pollutant_id", values = "normalised").fillna(0)
pivot_df.head(2)
```

```python
pivot_df.plot(kind = "area", stacked = "True", figsize = (8, 6))
plt.ylabel("Proportion of total pollution")
plt.title("Pollutant composition across states")
plt.legend(title = "Pollutant", bbox_to_anchor=(1.05, 1))
plt.xticks(rotation = 30)
plt.tight_layout()
```
